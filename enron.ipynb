```python
class Info:
    def __init__(self, val, text):
        self.val = val
        self.text = text

    def __gt__(self, other):        
        return self.val > other.val

    def __lt__(self, other):
        return self.val < other.val

    def __repr__(self):
        return "<Info({0}, {1})>".format(self.val, self.text)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# remove irrelevant words from frequency
ignore = nltk.corpus.stopwords.words("english")
ignore.append('ect')
ignore.append('enron')
ignore.append('k')
ignore.append('com')
ignore.append('e')
ignore.append('hou')
ignore.extend([w.lower() for w in nltk.corpus.names.words()])

def skip_unwanted(pos_tuple):
    word, tag = pos_tuple
    if not word.isalpha() or word in ignore:
        return False
    if tag.startswith("NN"):
        return False
    return True

positive_words = [word for word, tag in filter(
    skip_unwanted,
    nltk.pos_tag(nltk.corpus.movie_reviews.words(categories=["pos"]))
)]
negative_words = [word for word, tag in filter(
    skip_unwanted,
    nltk.pos_tag(nltk.corpus.movie_reviews.words(categories=["neg"]))
)]
```

```python
import os
from email.parser import Parser
from sortedcontainers import SortedList

import re

path = "./"
scorespp = [SortedList() for person in os.listdir(path) if person]
scores = SortedList()
master_words = dict()

for i, person in enumerate(os.listdir(path)):
    if person[0] != '.':
        sent = 'sent'
        master_words[person] = []
        print('i=',i)
        try:
            print(path + person + '/' + sent)
            for email in os.listdir(path + person + '/' + sent):
                words = []
                
                # get each email a person has sent
                name = (path + person + '/' + sent + '/' + email)
                text_file = open(name, "r")
                text = text_file.read()
                text_file.close()
                email = Parser().parsestr(text)
                
                body = email.get_payload()                
                
                #enforce at least 10 words
                if len(body.split(' ')) < 10:
                    continue
                
                # remove special characters and make all lower case
                body = re.sub('[^A-Za-z ]+', ' ', body.lower())
                
                #get individual words
                words = word_tokenize(body)
                
                #rank the email
                score = sentiment_scores(body)
                
                #make the Info object for sorted storage
                curinfo = Info(score['neg'], name)
                scorespp[int(i)].add(curinfo)
                scores.add(curinfo)
                
                # add current words to master list
                for j in words:
                    master_words[person].append(j)
                    
            #filter irrelevant words out
            filtered = [word for word in master_words[person] if word not in ignore]
            master_words[person] = filtered
            
            #check most frequenct words by person
            frequency = nltk.FreqDist(master_words[person])
            frequency.plot(40)
            
        except Exception as e:
            print(e)
                    
                    
                    
```
